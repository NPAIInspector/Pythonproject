```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
score = pd.read_csv('score2.csv', encoding='utf-8')
```

```python
model = KMeans(n_clusters=3).fit(score[['English', 'Japanese']])
print(model.labels_)
```

```python
score['cluster'] = model.labels_
score.to_csv('score4.csv')
score
```

```python
score.plot.scatter('English', 'Japanese', s=80, c=score['cluster'], edgecolor='black', figsize=(4, 4))

plt.savefig('scatter2.png')
plt.show()
```
